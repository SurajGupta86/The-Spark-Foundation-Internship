```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\Suraj Gupta\Downloads\SampleSuperstore.csv')
df.head()
```

```python
df.drop(['Country'],axis=1,inplace=True)
df
```

```python
df['Ship Mode'].value_counts()
```

```python
#TOTAL PROFIT AND SALES PER SUB-CATEGORY

plt.figure(figsize=(45,25))
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('Total Profit and Sales per Sub-Category')
plt.show()
```

```python
# COUNT OF SUB CATEGORY REGION WISE

plt.figure(figsize=(15,10))
sns.countplot(x="Sub-Category", hue="Region", data=df)
plt.show()
```

```python
#DISTRIBUTION OF PRODUCTS

plt.figure(figsize=(15,15))
df['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
df.groupby('Category')['City'].count()
```

```python
sns.catplot(x="Category", hue="Region", kind="count",data=df);
```

```python
sns.scatterplot("Sales",'Profit',data=df)
```

```python
pd.crosstab(df["Quantity"],df["Category"],df["Profit"],aggfunc='sum').plot(kind="bar",stacked=True,figsize=(15,10))
```
